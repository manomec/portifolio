```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Criando dados de exemplo
x = np.array([1, 2, 3, 4, 5])  # Variável independente
y = np.array([2, 4, 5, 4, 5])  # Variável dependente

# Criando um DataFrame
df = pd.DataFrame({'x': x, 'y': y})
```

```python
# Visualizando os dados
plt.scatter(df['x'], df['y'])
plt.xlabel('Variável Independente (x)')
plt.ylabel('Variável Dependente (y)')
plt.title('Relação entre x e y')
plt.show()
```

```python
# Adicionando uma constante para o termo de interceptação
X = sm.add_constant(df['x'])

# Ajustando o modelo OLS
model = sm.OLS(df['y'], X).fit()

# Imprimindo o resumo do modelo
print(model.summary())
```

```python
# Obter os coeficientes do modelo
model.params
```

```python
y_pred = model.predict(X)
```

```python
# Plotando a reta ajustada
plt.plot(x, y_pred, color='red', label='Reta ajustada')
plt.text(x.min(), y.min(), f'y = {model.params.x:.2f}x + {model.params.const:.2f}', fontsize=12, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regressão Linear com Statsmodels e Matplotlib')
plt.legend()
plt.show()
```
